# date time conversion-string to datetime

- %m|month of year 1-12
- %Y|year
- %H|hours 00-23
- %M|minutes 00-59
- %S|seconds 00-59

# immigration data analysis

```python
df=pd.read_excel('Canada.xlsx',sheet_name=1,skiprows=20,skipfooter=2)
df.head(2)
```

- rename columns
- drop columns
- add columns

```python
# rename
cols_to_rename={
    'OdName':'Country',
    'AreaName':"Continent",
    'REG':'Region',
    'DEV':'Status'
    }
df=df.rename(columns=cols_to_rename)
df.head(2)

```

```python
#drop
print('Type unique values:',df.Type.nunique())
print('Coverage unique values:',df.Coverage.nunique())

#count unique values
print('Coverage unique values\n',df.Coverage.value_counts())
```

```python
cols_to_drop=['AREA','Type','Coverage']
df=df.drop(columns=cols_to_drop)
df.head(2)
```

```python
#set index-not for every dataframe
df=df.set_index('Country')
df.head()
```

make all columns to lower case-optional

```python
df.columns=[str(name).lower() for name in df.columns.tolist()]
df.head()
```

```python
df
```

```python
years=list(map(str,range(1980,2014)))
print(years)
```

adding a column(total)

```python
df[years].sum(axis=1)# axis=1 for row wise sum,,,default is 0 (column wise sum)
```

```python
df['total']=df[years].sum(axis=1)
df.head()
```

```python
df.isnull().sum().plot(figsize=(10,1),kind='bar')
#no null value in dataset
```

total immigration visualization

```python
df.total.plot(rot=90,figsize=(25,5),xticks=range(0,df.shape[0],10))
```

```python
# top 25 countries with highest immigration
df.sort_values('total',ascending=False).head(25).plot(kind='bar',y='total',figsize=(30,10),
                                                      alpha=.5,
                                                      xlabel='Country',
                                                      ylabel="Immigration")
```
