slicing in a 2d matrix
```python
var[start: stop: step]  #1D

var[rstart:rstop:rstep,cstart:cstop:cstep]#2D

````python
import numpy as np
````

````python
x=np.random.randint(1,10,(5,5))
print(x)
````

````python
x[:,:2]
````

````python
x[:,-1]
````

````python
x[:,[0,1,-1]] #selective column index
````

````python
x[1:-1,[1,2,3]] #or
x[1:-1,1:-1]
````

boolean filtering




extract data from matrix based on condition

````python
x[x>5 ]#boolean matrix
````

````python
y=x[x<=5]
print(y.shape)
z=y.reshape(2,7)
print(z)
````

visualizing numerical data

````python
import matplotlib.pyplot as plt
````

````python

plt.plot([1,2,3],[4,5,6],linewidth=10,color='violet',linestyle='dashdot')
plt.plot([1,2,3],[6,5,4],linewidth=10,color='purple')
plt.xticks([1,2,3],rotation=90)
plt.yticks([4,5,6])
plt.title('first')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.text(x=2,y=5.2,s='cross region',fontsize=12,color='lightblue')
plt.show()
````
