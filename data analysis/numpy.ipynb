visualizing numerical data

````python
import matplotlib.pyplot as plt
````

````python

plt.plot([1,2,3],[4,5,6],linewidth=10,color='violet',linestyle='dashdot')
plt.plot([1,2,3],[6,5,4],linewidth=10,color='purple')
plt.xticks([1,2,3],rotation=90)
plt.yticks([4,5,6])
plt.title('first')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.text(x=2,y=5.2,s='cross region',fontsize=12,color='blue')
plt.show()
````

standard way of visualization

````python
fig,ax=plt.subplots(nrows=1,ncols=2,figsize=(10,2))
t=np.arange(-10,10,.1)
#print(t)
sine=np.sin(t)
cosine=np.cos(t)
#print(sine)
#print(cosine)
ax[0].plot(t,sine,label='sine',color='red',linestyle='',marker='x')
ax[1].plot(t,cosine,label='cosine',color='blue')
plt.savefig('graph',dpi=300)
````

Graphs




plt.plot=>line graph





plt.bar=>bar graph




plt.pie=>pie graph




plt.scatter=>scatter plot




plt.hist=>histogram

````python
fig,((g1,g2),(g3,g4))=plt.subplots(nrows=2,ncols=2,figsize=(10,10))
days=['sun','mon','tue','wed','thu','fri','sat']
sales=[160,150,140,145,156,123,152]
profit=np.random.normal(50,20,100)
g1.bar(days,sales)
g2.hist(profit,bins=[0,20,40,60,80,100],color='pink',edgecolor='black')
g3.pie(sales,labels=days,autopct='%.1f%%',explode=(0,0,0,0,0.1,0,0),wedgeprops={'width':0.7})
x=np.random.normal(0,1,500)
y=np.random.normal(0,1,500)
g4.scatter(x,y,color='purple',alpha=0.5)
plt.show()
````
